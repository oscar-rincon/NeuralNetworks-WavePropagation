```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
 
# Cargar los datos del archivo .txt
data = np.loadtxt('output/pressure_future_1000.txt')
# Extraer las coordenadas x, y y los valores de presión
x = data[:, 0]
y = data[:, 1]
pressure = data[:, 2]

Z=pressure.reshape((len(np.unique(y)), len(np.unique(x))))

# Crear una figura y un conjunto de ejes
plt.set_cmap('bwr_r') # a good start: blue to white to red colormap
# define your scale, with white at zero
vmin = -0.05 
vmax = 0.25
norm = colors.TwoSlopeNorm(vmin=vmin, vcenter=0, vmax=vmax)


# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots(figsize=(5, 5))

# Encontrar los valores mínimos y máximos excluyendo ceros
Z_nonzero = Z[np.nonzero(Z)]
min_val = np.min(Z_nonzero)
max_val = np.max(Z_nonzero)

 

# Graficar la presión como una imagen con ceros en blanco
pressure_image = ax.imshow(Z, vmax=max_val,vmin=min_val, cmap='RdBu')
 
# Agregar barra de color
cbar = fig.colorbar(pressure_image, ax=ax)
cbar.set_label('Pressure')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Guardar la gráfica como una imagen
plt.savefig('pressure_plot.png')

# Mostrar la gráfica
plt.show()
```

```python
# Crear una figura y un conjunto de ejes para la gráfica de línea
fig, ax = plt.subplots(figsize=(5, 3))

# Encontrar el índice del punto x=0 en el arreglo x
idx_x0 = 1000

# Extraer la sección transversal de los datos de presión en x=0
pressure_section = Z[:, idx_x0]

# Graficar la sección transversal de los datos de presión
ax.plot(pressure_section, color='blue')
ax.axhline(0, color='black', linestyle='--')  # Agregar una línea horizontal en y=0

# Etiquetas de los ejes
ax.set_xlabel('Y')
ax.set_ylabel('Pressure')

# Guardar la gráfica de línea como una imagen
plt.savefig('pressure_section_plot.png')

# Mostrar la gráfica de línea
plt.show()
```
