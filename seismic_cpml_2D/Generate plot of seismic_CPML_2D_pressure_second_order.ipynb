```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo .txt
data = np.loadtxt('output/pressure_future_1250.txt')
# Extraer las coordenadas x, y y los valores de presin
x = data[:, 0]
y = data[:, 1]
pressure = data[:, 2]

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Graficar la presin como una imagen
pressure_image = ax.imshow(pressure.reshape((len(np.unique(y)), len(np.unique(x)))),
                           extent=[min(x), max(x), min(y), max(y)], cmap='coolwarm_r', origin='lower')

# Agregar barra de color
cbar = plt.colorbar(pressure_image, ax=ax)
cbar.set_label('Pressure')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Guardar la grfica como una imagen
plt.savefig('pressure_plot.png')

# Mostrar la grfica
plt.show()
```
